```python
# Forgot Cov was a defined function in the text
def Cov(xs, ys, meanx = None, meany=None):
    xs=np.asarray(xs)
    ys=np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)

    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
def SpearmanCorr(xs, ys):
    xranks = pandas.Series(xs).rank()
    yranks = pandas.Series(ys).rank()
    return Corr(xranks, yranks)
```

```python
def EstimateS2(n=7, m = 1000):
    mu = 0
    sigma = 1
    estimates1 = []
    estimates2 = []
    for _ in range(m):
        xs = [random.gauss(mu, sigma) for i in range(n)]
        biased = np.var(xs)
        unbiased = np.var(xs, ddof=1)
        estimates1.append(biased)
        estimates2.append(unbiased)

    print('rmse biased', RMSE(estimates1, sigma**2))
    print('rmse unbiased', RMSE(estimates2, sigma**2))
```

```python
def RMSE(estimates, actual):
    e2 = [(estimate - actual)**2 for estimate in estimates]
    mse = np.mean(e2)
    return math.sqrt(mse)
```

```python
def SimulateSample(lam = 2, n = 10, m = 1000):
    estimates = []

    for j in range(m):
        xs = np.random.exponential(1/lam,n)
        lambdasample = 1/np.mean(xs)
        estimates.append(lambdasample)

    standarderror = RMSE(estimates, lam)
    print('Standard Error', standarderror)

    cdf = thinkstats2.Cdf(estimates)
    ci = cdf.Percentile(5), cdf.Percentile(95)
    print('Confidence Interval', ci)

    thinkplot.Cdf(cdf)
```

```python
SimulateSample(lam = 2, n = 10, m = 1000)
```

```python
SimulateSample(lam = 2, n = 100, m = 1000)
```

```python
SimulateSample(lam = 2, n = 500, m = 1000)
```

```python
SimulateSample(lam = 2, n = 1500, m = 1000)
```
